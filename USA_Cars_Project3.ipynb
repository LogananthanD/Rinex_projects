# **Challenge_17**

```python
# What are the top 5 most expensive car brands on average?

expensive_brands = df.groupby('brand')['price'].mean().nlargest(5)

print("Top 5 most expensive car brands on average:")
for brand, average_price in expensive_brands.items():
    print(f"- {brand}: ${average_price:.2f}")
```

# **Challenge_20**

```python
# What is the average price of cars for each color?

average_price_by_color = df.groupby('color')['price'].mean()

print("Average price by color:")
for color, average_price in average_price_by_color.items():
    print(f"- {color}: ${average_price:.2f}")
```

# **Challenge_21**

```python
# Can you classify cars into different price categories (e.g., budget, mid-range, luxury) based on their prices?

def classify_cars_by_price(price):
  if price < 10000:
    return "Budget"
  elif price < 30000:
    return "Mid-range"
  else:
    return "Luxury"

df["price_category"] = df["price"].apply(classify_cars_by_price)

price_category_counts = df["price_category"].value_counts()
print(price_category_counts)
```

# **Challenge_22**

```python
# How do car prices vary between different title statuses (e.g., clean, salvage)?

average_price_by_title_status = df.groupby("title_status")["price"].mean()

print("Average price by title status:")
for title_status, average_price in average_price_by_title_status.items():
    print(f"- {title_status}: ${average_price:.2f}")
```

# **Challenge_23**

```python
# What is the distribution of car prices in the dataset?

import matplotlib.pyplot as mpt

prices = df["price"]

mpt.hist(prices, bins=10, edgecolor="black")
mpt.xlabel("Price")
mpt.ylabel("Number of Cars")
mpt.title("Distribution of Car Prices")
mpt.show()
```

# **Challenge_24**

```python
# Can you visualize the distribution of car conditions (e.g., new, used, like new) in a bar chart?

import matplotlib.pyplot as mpt

condition_counts = df['condition'].value_counts()

mpt.bar(condition_counts.index, condition_counts.values)
mpt.xlabel("Condition")
mpt.ylabel("Number of Cars")
mpt.title("Distribution of Car Conditions")
mpt.show()
```

# **Challenge_25**

```python
# How do the average prices of cars manufactured in each year vary over time? Create a line plot to visualize this trend.

average_prices_by_year = df.groupby('year')['price'].mean()

years = average_prices_by_year.index
prices = average_prices_by_year.values

mpt.plot(years, prices)
mpt.xlabel("Year")
mpt.ylabel("Average Price")
mpt.title("Average Car Prices by Year")
mpt.show()
```

# **Challenge_26**

```python
# Can you visualize the distribution of car colors using a pie chart?

color_counts = df['color'].value_counts()

colors = color_counts.index
counts = color_counts.values

mpt.pie(counts, labels=colors, autopct="%1.1f%%")
mpt.title("Distribution of Car Colors")
mpt.show()
```
